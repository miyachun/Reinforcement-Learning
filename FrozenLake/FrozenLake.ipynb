```python
n_t=20
```

```python
s=env.reset()
for t in range(n_t):
  a=env.action_space.sample()
  s1,r,done,_,_=env.step(a)
  print('t=',t,'s=',s,'a=',a,'r=',r,'s\'=',s1,'done=',done)
  s=s1
  if done:
    break
```

```python
import time
from IPython.display import clear_output

def game_display(img):
  plt.imshow(img)
  plt.show()
  time.sleep(0.5)
  clear_output(wait=True)
```

```python
env=gym.make('FrozenLake-v1', render_mode='rgb_array', is_slippery=False)
```

```python
n_t=20
```

```python
s=env.reset()
game_display(env.render())
for t in range(n_t):
  a=env.action_space.sample()
  s1,r,done,_,_=env.step(a)
  print('t=',t,'s=',s,'a=',a,'r=',r,'s\'=',s1,'done=',done)
  game_display(env.render())
  s=s1
  if done:
    break
```
